```python
import numpy as np
import pandas as pd

from rice_ml.supervised_learning import MLPBinaryClassifier

df = pd.read_csv("../data/lesions_processed.csv")

X = df[["x_norm","y_norm","slice_norm","r_xy","experiment"]].values
y = df["tampered"].values.astype(int)
```

```python
#train
rng = np.random.default_rng(42)
idx = np.arange(len(y))
rng.shuffle(idx)
split = int(0.75 * len(y))
tr, te = idx[:split], idx[split:]

Xtr, Xte = X[tr], X[te]
ytr, yte = y[tr], y[te]
```

```python
#fit
mlp = MLPBinaryClassifier(hidden_units=8, lr=0.3, n_epochs=400, random_state=42)
mlp.fit(Xtr, ytr)

pred = mlp.predict(Xte)
acc = np.mean(pred == yte)
acc
```
