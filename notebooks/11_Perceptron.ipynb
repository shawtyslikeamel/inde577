```python
import numpy as np
import pandas as pd

from rice_ml.supervised_learning import Perceptron

df = pd.read_csv("../data/lesions_processed.csv")

X = df[["x_norm","y_norm","slice_norm","r_xy","experiment"]].values
y = df["tampered"].values.astype(int)
```

```python
#train/test split
rng = np.random.default_rng(42)
idx = np.arange(len(y))
rng.shuffle(idx)
split = int(0.75 * len(y))
tr, te = idx[:split], idx[split:]

Xtr, Xte = X[tr], X[te]
ytr, yte = y[tr], y[te]
```

```python
#fit
model = Perceptron(lr=0.1, n_epochs=50, random_state=42)
model.fit(Xtr, ytr)

pred = model.predict(Xte)
acc = np.mean(pred == yte)
acc
```

The perceptron is a linear classifier trained with an online update rule. It predicts using a step function on a linear score and updates weights only when samples are misclassified. This is a foundational neural-network model.
